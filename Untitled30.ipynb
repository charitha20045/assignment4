```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = sns.load_dataset('tips')
print(df.head())
df.hist(figsize=(10, 8), bins=20)
plt.suptitle('Histograms of Tips Dataset Features', fontsize=16)
plt.show()

# Boxplots
plt.figure(figsize=(10, 6))
sns.boxplot(data=df.select_dtypes(include=['float64', 'int64']))
plt.title('Boxplot of Tips Dataset Features')
plt.show()
sns.pairplot(df)
plt.suptitle('Pairplot of Tips Dataset Features', y=1.02, fontsize=16)
plt.show()
plt.figure(figsize=(8, 6))
corr = df.select_dtypes(include=['float64', 'int64']).corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap of Tips Dataset')
plt.show()
```
